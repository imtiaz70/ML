```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# import pylab as pl
%matplotlib inline
df = pd.read_csv("FuelConsumption.csv")
df.head()
cfd = df[['ENGINESIZE','FUELCONSUMPTION_COMB','CO2EMISSIONS','CYLINDERS']]
cfd.hist()
plt.show()
plt.tight_layout()
cfd.head()
```

### Traning & Testing

```python
msk = np.random.rand(len(df))<0.8
train = cfd[msk]
test = cfd[~msk]
```

### Traning Data Distribution

```python
plt.scatter(cfd.ENGINESIZE , cfd.CO2EMISSIONS)
```

```python
plt.scatter(train.ENGINESIZE , train.CO2EMISSIONS)
plt.xlabel("Engin Size")
plt.ylabel("Emisssion")
plt.show()
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_X = np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])
regr.fit(train_X,train_y)
# COEFFICIENTS
print('Coefficients: ' , regr.coef_)
print('intersept' , regr.intercept_)
```

```python
plt.scatter(train.ENGINESIZE , train.CO2EMISSIONS , color = 'blue')
plt.plot(train_X, regr.coef_*train_X + regr.intercept_[0],'-r')
plt.xlabel("Engin Size")
plt.ylabel("Emission")
```
